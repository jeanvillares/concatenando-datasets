```python
# Biblioteca para manipulação e análise de dados
import pandas as pd

# Biblioteca para criar listas de arquivos buscando-os em diretórios
import glob
```

```python
# A variável 'files' recebe uma lista ordenada dos arquivos que serão concatenados
# A parte legal é que, com o glob, se há um padrão no arquivo, podemos simplesmente utilizar o caractere '*'. Para este caso
# o arquivo possui os nomes METADATA1, METADATA2, METADATA3, e assim por diante. Assim somente os arquivos com o mesmo padrão
# são importados.

files = sorted(glob.glob('xlsfiles/DADOS-*.xls')) 
files[:5]
```

```python
# Os arquivos possuem o mesmo número e nomes de colunas. Dessa maneira é possível concatená-los, o que torna
# o trabalho mais simples. Através de um loop for, os arquivos são atribuídos a uma variável 'file' e lida pelo 'pd.read_excel'
# ignorando o cabeçalho.

df = pd.concat((pd.read_excel(file).assign(filename = file) for file in files), ignore_index = True)
```
